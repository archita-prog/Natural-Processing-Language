```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
#sentence tokenization
doc = nlp("I like sushi. My friend likes pizza.")
for sentence in doc.sents:
    print(sentence)
```

```python
#word tokenization
for sentence in doc.sents:
    for word in sentence:
        print(word)
```
